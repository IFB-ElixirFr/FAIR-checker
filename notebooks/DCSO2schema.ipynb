```python
kg = ConjunctiveGraph()
```

```python
kg.parse("https://github.com/RDA-DMP-Common/RDA-DMP-Common-Standard/raw/master/ontologies/examples/dcso.example.4.0.0.ttl", format="turtle")
#kg.parse("dcso.example.4.0.0.ttl", format="turtle")
```

```python
print(f"Loaded {len(kg)} triples")
out = kg.serialize(format="turtle")

print(out)
```

```python
prefix = """
PREFIX dcso: <https://w3id.org/dcso/ns/core#> 
PREFIX schema: <https://schema.org/>
PREFIX dct: <http://purl.org/dc/terms/> 
"""

dmp_construct_query = prefix + """

CONSTRUCT {

    ?s rdf:type schema:CreativeWork ;
        schema:title ?title ;
        schema:description ?description ;
        schema:dateModified ?modified ; 
        schema:dateCreated ?created ;
        dcso:ethicalIssuesExist ?eth ;
        schema:accountablePerson ?contact ; 
        schema:hasPart ?dataset ; 
        schema:identifier ?dmp_id ; 
        schema:inLanguage ?language . 

} WHERE {

    ?s rdf:type dcso:DMP ;
       dct:title ?title ;

       dcso:created ?created ;
       dcso:modified ?modified ;
       dcso:ethicalIssuesExist ?eth ;
       dcso:hasContact ?contact ; 
       dcso:hasDataset ?dataset ; 
       dcso:hasDMPId ?dmp_id ; 
       dcso:language ?language .

}

"""
```

```python
kg_res = kg.query(dmp_construct_query)
out = kg_res.serialize(format="turtle")
print(out.decode())

#out = kg_res.serialize(format="json-ld")
#print(out.decode())
```

```python
prefix = """
PREFIX dcso: <https://w3id.org/dcso/ns/core#> 
PREFIX schema: <https://schema.org/>
PREFIX dct: <http://purl.org/dc/terms/> 
"""

dataset_construct_query = prefix + """

CONSTRUCT {

    ?s rdf:type schema:Dataset ;
        schema:identifier ?dataset_id ;
        schema:title ?title .

} WHERE {

    ?s rdf:type dcso:Dataset ;
       dcso:sensitiveData ?sensitive ;
       dcso:hasDatasetId ?dataset_id .

    OPTIONAL {?s dct:title ?title .}
}

"""
```

```python
kg_res = kg.query(dataset_construct_query)
out = kg_res.serialize(format="turtle")
print(out.decode())

#out = kg_res.serialize(format="json-ld")
#print(out.decode())
```

```python
prefix = """
PREFIX dcso: <https://w3id.org/dcso/ns/core#> 
PREFIX schema: <https://schema.org/>
PREFIX dct: <http://purl.org/dc/terms/> 
PREFIX foaf: <http://xmlns.com/foaf/0.1/>
"""

contact_construct_query = prefix + """

CONSTRUCT {

    ?s rdf:type schema:Person ;
        schema:identifier ?contact_id ;
        schema:givenName ?name ;
        schema:email ?email .

} WHERE {

    ?s rdf:type dcso:Contact ;
       dcso:hasContactId ?contact_id ;
       foaf:mbox ?email ;
       foaf:name ?name .
}

"""
```

```python
kg_res = kg.query(contact_construct_query)
out = kg_res.serialize(format="turtle")
print(out.decode())
```
