```python
import requests
from tqdm.notebook import tqdm
import json
from rdflib import ConjunctiveGraph
```

```python
for doi in samples_dois_list:
    print(doi)
```

```python
print(len(samples_dois_list))
```

```python
for url in tqdm(samples_dois_list):
    response = requests.get("http://0.0.0.0:5000/api/inspect/get_rdf_metadata/" + url)
    graph = json.dumps(response.json(), ensure_ascii=False)
    kg = ConjunctiveGraph()
    kg = kg.parse(data=graph, format="json-ld")
    print("CODE: %s - Len KG: %s - URL: %s" % (response.status_code, len(kg), url))
```
